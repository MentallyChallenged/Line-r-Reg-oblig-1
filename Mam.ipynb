```python

import numpy as np
import matplotlib.pyplot as plt

xData = np.array([0,0.3142,0.6283,0.9425,1.2566,1.5708,1.8850,2.1991,2.5133,
                  2.8274,3.1416,3.4558,3.7699,4.0841,4.3982,4.7124,5.0265,5.3407,
                  5.6549,5.9690,6.2832])
yData = np.array([1.1230,3.6365,2.6662,0.0671,- 1.6282,0.0434,2.2371,3.4170,
                  2.6144,-0.2578,- 0.9083,1.5468,2.8202,3.4206,1.2458,-1.0849,
                  - 0.2882,1.5803,3.0363,2.3500,0.1370])

def F(param):
    A, B, C = param
    return np.sum((A + B * np.sin(C * xData) - yData) ** 2)

def gradient(F, param, h=1e-6):
    grad = np.zeros(len(param))
    for i in range(len(param)):
        param_fremover = param.copy()
        param_fremover[i] += h
        grad[i] = (F(param_fremover) - F(param)) / h
    return grad

def descent(F, gradient, param, gamma=1e-5, kmax=10000, tol=1e-6):
    for k in range(1, kmax):
        grad = gradient(F, param)
        param -= gamma * grad
        if np.linalg.norm(grad) < tol:
            break
    return param

A0, B0, C0 = 3, 3, 3
gamma = 1e-5
param_iv = np.array([A0, B0, C0], dtype=np.float64)

A_opt, B_opt, C_opt = descent(F, gradient, param_iv)


x_fit = np.linspace(0, np.max(xData), 500)
y_fit = A_opt + B_opt * np.sin(C_opt * x_fit)

plt.figure(figsize=(10, 6))
plt.scatter(xData, yData, color='blue', label='Datapunkter', marker='o', edgecolors='black', s=60, alpha=0.8)
plt.plot(x_fit, y_fit, color='red', label='Tilpasset funksjon', linewidth=2)
plt.grid(True)
plt.xlabel("x", fontsize=20)
plt.ylabel("y", fontsize=20)
plt.legend(fontsize=12, loc='best')
plt.title("Optimalisering av sin-funksjon", fontsize=20)
plt.tight_layout()
plt.show()

print(f"Estimerte verdier med 𝛾 = {gamma}: ")
print(f"A = {A_opt:.3f}")
print(f"B = {B_opt:.3f}")
print(f"C = {C_opt:.3f}")
```

$h(x, y) = x^2 + y^2 + 2(1 − 0.2x) cos(3x)(0.5 + sin(3y))$

for x ∈ [−2, 2] og y ∈ [−2, 2]. Det er mange lokale minima, utfordringa er å bestemme det globale. Dere skal/kan bruke skriptet GradDescnD.py som ligger på Canvas.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def h(x, y):
    return x**2 + y**2 + 2*(1 - 0.2*x) * np.cos(3*x) * (0.5 + np.sin(3*y))


def grad_h(x, y):
    dh_dx = 2*x - 0.4*np.cos(3*x)*(0.5 + np.sin(3*y)) - 6*(1 - 0.2*x)*np.sin(3*x)*(0.5 + np.sin(3*y))
    dh_dy = 2*y + 6*(1 - 0.2*x)*np.cos(3*x)*np.cos(3*y)
    return np.array([dh_dx, dh_dy])


def gradient_descent(start, learning_rate, num_iterations=100):
    x, y = start
    path = [(x, y)]
    for _ in range(num_iterations):
        grad = grad_h(x, y)
        x -= learning_rate * grad[0]
        y -= learning_rate * grad[1]
        path.append((x, y))
    return np.array(path)


def gradient_descent_decay(start, gamma_0, num_iterations=100):
    x, y = start
    path = [(x, y)]
    for n in range(num_iterations):
        gamma_n = gamma_0 * np.exp(-0.1 * n)
        grad = grad_h(x, y)
        x -= gamma_n * grad[0]
        y -= gamma_n * grad[1]
        path.append((x, y))
    return np.array(path)


x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = h(X, Y)


fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('h(x, y)')
ax.set_title('3D-plott av funksjonen h(x, y)')
plt.show()


fig, ax = plt.subplots(figsize=(8, 6))
contour = ax.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Konturplott av h(x, y)')
plt.show()

# Test gradient descent med fast læringsrate
start_points = [(-1.5, 1.5), (1.5, -1.5), (0, 0), (-1, -1)]
learning_rates = [0.01, 0.1, 0.2]

fig, ax = plt.subplots(figsize=(8, 6))
contour = ax.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Gradient descent med ulike startpunkter')

for start in start_points:
    for lr in learning_rates:
        path = gradient_descent(start, lr)
        ax.plot(path[:, 0], path[:, 1], marker='o', markersize=2, label=f'Start {start}, lr={lr}')

ax.legend()
plt.show()


fig, ax = plt.subplots(figsize=(8, 6))
contour = ax.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Gradient descent med avtagende læringsrate')

for start in start_points:
    path = gradient_descent_decay(start, gamma_0=0.2)
    ax.plot(path[:, 0], path[:, 1], marker='o', markersize=2, label=f'Start {start}')

ax.legend()
plt.show()
```

## Del 5 Mehdi

I sekantmetoden brukes algoritmen
\begin{equation}
x_{n+1} = x_n - f(x_n)\frac{(x_n-x_{n-1})}{f(x_n)-f(x_{n-1})}  , 
\end{equation}
med n = 1,2,3...
og startpunkter x0 og x1 til å beregne approksimasjoner {xn} til nullpunktet til funksjonen f . Vi antar at følgen {xn} konvergerer mot nullpunktet x∗. Spørsmålet i denne oppgaven er: “Hvor rask er konvergensen”? Dette spørsmålet skal besvares ved at størrelsene C og p i approksimasjonen

$|x_{n+1} - x^*| \approx C|x_{n} - x^*|^p$

skal bestemmes. Dette er atskillig mer utfordrende å gjøre for sekantmetoden enn  for Newtons metode. Med utgangspunkt i formelen for sekantmetoden og definisjonen av avviket en som e_n = xn − x∗ slik at xn = x∗ + en får vi følgende ligning for avviket en+1 i iterasjon n + 1:

$e{n+1} = e_{n} - f(x^* + e_{n})\frac{e_{n}-e_{n-1}}{f(x^*+e_{n})-f(x^+e_{n-1})}$

Dersom vi antar at f er to ganger deriverbar og at ε er et lite tall, er \ tilnærminga

$f(x^* + \epsilon) \approx f(x^*) + f'(x^*)\epsilon + \frac{1}{2}f''(x^*)\epsilon^2 = f'(x^*)\epsilon +\frac{1}{2}f''(x^*) \epsilon^2$  

gyldig, dvs. at vi kan uttrykke $f(x^*+\epsilon) \approx \epsilon f'(x^*)(1+M\epsilon)$, hvor $M = \frac{f''(x^*)}{(2f'(x^*))}$  og videre, siden en og en+1 antas små, er også approksimasjonen nedenfor gyldig. 

$1 \approx \frac{1}{1+M(e{n}+e{n+1})}$

a) Bruk ligningene og approksimasjonene til å vise 
\begin{equation}
e_{n+1} \approx M e_n e_{n-1}
\end{equation}
hvor

\begin{equation}
M = \frac{f''(x^*)}{2f'(x^*)}
\end{equation}

Vi starter vi med iterasjonsformeln:

\begin{equation}
e_{n+1} = e_n - \frac{f(x^* + e_n) (e_n - e_{n-1})}{f(x^* + e_n) - f(x^* + e_{n-1})}
\end{equation}

Bruk Taylor-approksimasjonen for $f(x^* + e)$:

\begin{equation}
f(x^* + e) \approx f'(x^*) e + \frac{1}{2} f''(x^*) e^2
\end{equation}

Setter dette inn tilbake i telleren og nevneren:

Telleren blir da:

\begin{equation}
\frac{(f'(x^*) e_n + \frac{1}{2} f''(x^*) e_n^2)(e_n - e_{n-1})}{f(x^* + e_n) - f(x^* + e_{n-1})}
\end{equation}

Nevneren dermend blir:

\begin{equation}
\frac{(f'(x^*) e_n + \frac{1}{2} f''(x^*) e_n^2)(e_n - e_{n-1})}{(f'(x^*)(e_n - e_{n-1}) + \frac{1}{2} f''(x^*) (e_n - e_{n-1})(e_n + e_{n-1}))}
\end{equation}

Sett inn i iterasjonsformelen:

\begin{equation}
e_{n+1} = e_n - \frac{(e_n - e_{n-1}) f'(x^*) (e_n + M e_n^2)}{(e_n - e_{n-1}) f'(x^*) (1 + M (e_n + e_{n-1}))}
\end{equation}

Forkorter $(e_n - e_{n-1}) f'(x^*)$:

\begin{equation}
e_{n+1} = e_n - \frac{e_n + M e_n^2}{1 + M(e_n + e_{n-1})}
\end{equation}

For små $e_n$, bruker tilnærmingen:

\begin{equation}
\frac{1}{1+M(e_n+e_{n-1})} \approx 1 - M(e_n + e_{n-1})
\end{equation}

Så vi får:

\begin{equation}
e_{n+1} \approx e_n - (e_n + M e_n^2)(1 - M(e_n + e_{n-1}))
\end{equation}

Multipliser ut:

\begin{equation}
e_{n+1} \approx e_n - e_n + M e_n^2 + M e_n e_{n-1} - M^2 e_n^3 - M^2 e_n^2 e_{n-1}
\end{equation}

For små $e_n$ ignorer vi høyere ordens ledd og får igjen::

\begin{equation}
e_{n+1} \approx M e_n e_{n-1}
\end{equation}


hvor

\begin{equation}
M = \frac{f''(x^*)}{2f'(x^*)}
\end{equation}

## Del 5b
Vis ved å sette inn approksimasjonen $e_{n+1} \approx C e^p_n$ i resultatet fra a) at

\begin{equation}
e^{p-1}_n \approx \frac{M}{C}e_{n-1}
\end{equation}

og vi får 
\begin{equation}
C^{p-1}e^{p(p-1)}_{n-1}\approx \frac{M}{C}e_{n-1}
\end{equation}

fra forrige oppgfave har vi: $e_{n+1} =  Me_{n}e_{n-1}$ og $e_{n+1} \approx Ce_{n}^p$

Dette er like, så vi kan sette det sammen: $Me_{n}e_{n-1} = Ce_{n}^p$ , også deler vi begge sider på C, får å få en alene: 

\begin{equation}
\frac{M}{C}e_{n}e_{n-1} \approx e_{n}^p
\end{equation}

deler begge sider på e_n

\begin{equation}
\frac{M}{C}e_{n-1} \approx e_{n}^{p-1}

\begin{equation}

\end{equation}

\begin{equation}
C^{p-1}e_{n-1}^{p(p-1)} \approx Ce_{n-1}^p
\end{equation}
