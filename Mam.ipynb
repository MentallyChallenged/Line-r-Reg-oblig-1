```python
# Importer nødvendige biblioteker igjen etter reset
import numpy as np

# Definer funksjonen g(x) basert på brukerens spesifikasjon
def g_new(x):
    return 2 * (4*x - 2/5) * np.exp(-2*(x - 0.1)**2)

# Implementering av fikspunktiterasjon for å finne min-punktene
def fixed_point_iteration(g, x0, tol=1e-6, max_iter=100):
    """Fikspunktiterasjon for å finne x-koordinatene til min-punktene."""
    x = x0
    for _ in range(max_iter):
        x_new = g(x)
        if abs(x_new - x) < tol:  # Sjekker konvergens
            return x_new
        x = x_new
    return x  # Returnerer siste verdi hvis maks iterasjoner nås

# Velg startpunkter nær forventede min-punkter basert på plottet
x0_1 = -1.5  # Estimert venstre minimum
x0_2 = 1.5   # Estimert høyre minimum

# Finn fikspunktene numerisk
x_min1 = fixed_point_iteration(g_new, x0_1)
x_min2 = fixed_point_iteration(g_new, x0_2)

(x_min1, x_min2)
```

```python
# Implementering av éndimensjonal gradient descent for å finne det globale minpunktet

# Definer funksjonen f(x) og dens deriverte f'(x)
def f(x):
    return x**2 + np.exp(-2*(x - 0.1)**2)

def df(x):
    return 2*x - 4*(x - 0.1)*np.exp(-2*(x - 0.1)**2)

# Gradient descent-algoritme
def gradient_descent(df, x0, gamma, tol=1e-6, max_iter=1000):
    """Éndimensjonal gradient descent for å finne minimum av f(x)."""
    x = x0
    for _ in range(max_iter):
        x_new = x - gamma * df(x)
        if abs(x_new - x) < tol:  # Sjekker konvergens
            return x_new
        x = x_new
    return x  # Returnerer siste verdi hvis maks iterasjoner nås

# Test ulike verdier for læringsraten gamma
gammas = [0.01, 0.1, 0.5, 1.0]

# Startpunktene
x0_left = -2
x0_right = 2

# Kjøre gradient descent for ulike gamma-verdier
results = {}
for gamma in gammas:
    x_min_left = gradient_descent(df, x0_left, gamma)
    x_min_right = gradient_descent(df, x0_right, gamma)
    results[gamma] = (x_min_left, x_min_right)

results
```

I sekantmetoden brukes algoritmen

$x_{n+1} = x_n - f(x_n)\frac{(x_n-x_{n-1})}{f(x_n)-f(x_{n-1})}$  , med n = 1,2,3...

med startpunkter x0 og x1 til å beregne approksimasjoner {xn} til nullpunktet til funksjonen f . Vi antar at følgen {xn} konvergerer mot nullpunktet x∗. Spørsmålet i denne oppgaven er: “Hvor rask er konvergensen”? Dette spørsmålet skal besvares ved at størrelsene C og p i approksimasjonen

$|x_{n+1} - x^*| \approx C|x_{n} - x^*|^p$

skal bestemmes. Dette er atskillig mer utfordrende å gjøre for sekantmetoden enn  for Newtons metode. Med utgangspunkt i formelen for sekantmetoden og definisjonen av avviket en som e_n = xn − x∗ slik at xn = x∗ + en får vi følgende ligning for avviket en+1 i iterasjon n + 1:

$e{n+1} = e_{n} - f(x^* + e_{n})\frac{e_{n}-e_{n-1}}{f(x^*+e_{n})-f(x^+e_{n-1})}$

Dersom vi antar at f er to ganger deriverbar og at ε er et lite tall, er \ tilnærminga

$f(x^* + \epsilon) \approx f(x^*) + f'(x^*)\epsilon + \frac{1}{2}f''(x^*)\epsilon^2 = f'(x^*)\epsilon +\frac{1}{2}f''(x^*) \epsilon^2$  

gyldig, dvs. at vi kan uttrykke $f(x^*+\epsilon) \approx \epsilon f'(x^*)(1+M\epsilon)$, hvor $M = \frac{f''(x^*)}{(2f'(x^*))}$  og videre, siden en og en+1 antas små, er også approksimasjonen nedenfor gyldig. 

$1 \approx \frac{1}{1+M(e{n}+e{n+1})}$

a) Bruk ligningene og approksimasjonene til å vise 
\begin{equation}
e_{n+1} \approx M e_n e_{n-1}
\end{equation}
hvor

\begin{equation}
M = \frac{f''(x^*)}{2f'(x^*)}
\end{equation}

Vi starter vi med iterasjonsformeln:

\begin{equation}
e_{n+1} = e_n - \frac{f(x^* + e_n) (e_n - e_{n-1})}{f(x^* + e_n) - f(x^* + e_{n-1})}
\end{equation}

Bruk Taylor-approksimasjonen for \( f(x^* + e) \):

\begin{equation}
f(x^* + e) \approx f'(x^*) e + \frac{1}{2} f''(x^*) e^2
\end{equation}

Setter dette inn tilbake i telleren og nevneren:

Telleren blir da:

\begin{equation}
(f'(x^*) e_n + \frac{1}{2} f''(x^*) e_n^2)(e_n - e_{n-1})
\end{equation}

Nevneren dermend blir:

\begin{equation}
f'(x^*)(e_n - e_{n-1}) + \frac{1}{2} f''(x^*) (e_n - e_{n-1})(e_n + e_{n-1})
\end{equation}

Sett inn i iterasjonsformelen:

\begin{equation}
e_{n+1} = e_n - \frac{(e_n - e_{n-1}) f'(x^*) (e_n + M e_n^2)}{(e_n - e_{n-1}) f'(x^*) (1 + M (e_n + e_{n-1}))}
\end{equation}

Forkorter $(e_n - e_{n-1}) f'(x^*)$:

\begin{equation}
e_{n+1} = e_n - \frac{e_n + M e_n^2}{1 + M(e_n + e_{n-1})}
\end{equation}

For små $e_n$, bruker tilnærmingen:

\begin{equation}
\frac{1}{1+M(e_n+e_{n-1})} \approx 1 - M(e_n + e_{n-1})
\end{equation}

Så vi får:

\begin{equation}
e_{n+1} \approx e_n - (e_n + M e_n^2)(1 - M(e_n + e_{n-1}))
\end{equation}

Multipliser ut:

\begin{equation}
e_{n+1} \approx e_n - e_n + M e_n^2 + M e_n e_{n-1} - M^2 e_n^3 - M^2 e_n^2 e_{n-1}
\end{equation}

For små \( e_n \) ignorer vi høyere ordens ledd:

\begin{equation}
e_{n+1} \approx M e_n e_{n-1}
\end{equation}

Dermed har vi vist at:

\begin{equation}
e_{n+1} \approx M e_n e_{n-1}
\end{equation}

hvor

\begin{equation}
M = \frac{f''(x^*)}{2f'(x^*)}
\end{equation}
